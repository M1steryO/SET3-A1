# Построение графиков для задания A1


```python
import math
import matplotlib.pyplot as plt
import pandas as pd

exact_value = 0.25 * math.pi + 1.25 * math.asin(0.8) - 1


data_part1 = pd.read_csv('a1_1.txt', sep=" ", header=None, names=["points", "approximation"])
data_part2 = pd.read_csv('a1_2.txt', sep=" ", header=None, names=["points", "approximation"])


data_part1['relative_deviation'] = abs(data_part1['approximation'] - exact_value) / exact_value
data_part2['relative_deviation'] = abs(data_part2['approximation'] - exact_value) / exact_value
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(data_part1['points'], data_part1['approximation'], label='Приближенная площадь', color='b')
plt.axhline(exact_value, color='r', linestyle='--', label='Точное значение')
plt.xlabel('Число точек')
plt.ylabel('Приближенная площадь')
plt.title('Зависимость приближенной площади от числа точек')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(data_part1['points'], data_part1['relative_deviation'], label='Относительное отклонение', color='g')
plt.xlabel('Число точек')
plt.ylabel('Относительное отклонение')
plt.title('Зависимость относительного отклонения от числа точек')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(data_part2['points'], data_part2['approximation'], label='Приближенная площадь', color='b')
plt.axhline(exact_value, color='r', linestyle='--', label='Точное значение')
plt.xlabel('Число точек')
plt.ylabel('Приближенная площадь')
plt.title('Зависимость приближенной площади от числа точек')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(data_part2['points'], data_part2['relative_deviation'], label='Относительное отклонение', color='g')
plt.xlabel('Число точек')
plt.ylabel('Относительное отклонение')
plt.title('Зависимость относительного отклонения от числа точек')
plt.legend()
plt.tight_layout()
plt.show()
```

### Выводы

График приближенной площади показывает, что с увеличением числа точек оценка площади становится более стабильной и приближается к точному значению. В начале, при малом числе точек, наблюдаются значительные колебания в оценке площади, что говорит о недостаточной точности метода при таком количестве точек.

На графике относительного отклонения видно, что погрешность вычислений существенно уменьшается с увеличением числа точек. При малом числе точек относительное отклонение может достигать 5%, но затем оно стабилизируется и держится на уровне около 1-2%. После 60000 точек улучшение точности становится незначительным, что говорит о достижении оптимального соотношения между вычислительными затратами и точностью результата.

Практический вывод состоит в том, что для получения надежной оценки площади с относительной погрешностью около 1% достаточно использовать около 60000 точек. Дальнейшее увеличение числа точек приводит лишь к незначительному улучшению точности при существенном увеличении вычислительных затрат. Монотонное убывание относительного отклонения подтверждает корректность реализации метода Монте-Карло, так как это соответствует теоретическим ожиданиям о сходимости метода при увеличении числа испытаний.
